```python
%pylab inline
rcParams['figure.figsize'] = (14, 6) #wide graphs by default
from __future__ import print_function
from __future__ import division
import scipy.ndimage as ndimage
import numpy as np
```

```python
# read image and convert from 8 bit integer to float, *-1 to fix sign,
# then convert to grayscale

img = imread('gecko.jpg')
img = img.astype(float) * -1 + 255

img_gray=[3]
for i in range(1,3):
    img_gray += img[:,:,i]
img_gray = (img_gray/3) * -1 + 255

subplot(121)
imshow(img)
colorbar(shrink=.6)
subplot(122)
imshow(img_gray, cmap='gray')
colorbar(shrink=.6)
```

```python
# The hypot result of two sobel filters on the grayscale image
# Result is used as the axiom of the next cell's operation

src_image = img_gray
sx = ndimage.sobel(src_image, axis=0, mode='constant')
sy = ndimage.sobel(src_image, axis=1, mode='constant')

filtered_image = hypot(sx, sy)
imshow(filtered_image)
```

```python
# Hypot two sobel filters, save to disk as jpg, then compute new sobel filters from the result

number_of_frames = 240
src_image = filtered_image
first_frame = filtered_image
last_frame = []
sx = []
sy = []
animate = []
for i in range(1,number_of_frames):
    sx = ndimage.sobel(src_image, axis=0, mode='constant')
    sy = ndimage.sobel(src_image, axis=1, mode='constant')
    src_image = hypot(sx, sy)
    
    filename = "anim_frames/out_%03d.jpg" % i
    filename = "anim_frames/out_{0:03d}.jpg".format(i)
#     imsave(filename, src_image)
    print("Saved " + filename)
    
    if i == (number_of_frames-1):
        last_frame = src_image
```

```python
subplot(121)
title("First Frame")
imshow(first_frame)
subplot(122)
title("Last Frame")
imshow(last_frame)
```
