```python
%pylab inline
rcParams['figure.figsize'] = (14, 6) #wide graphs by default
from __future__ import print_function
from __future__ import division
import scipy.ndimage as ndimage
from PIL import Image, ImageSequence 
from images2gif import writeGif
 
lena = lena()
```

```python
# read image and convert from 8 bit integer to float, *-1 to fix sign
img = imread('gecko.jpg')
img = img.astype(float) * -1 + 255
imshow(img)
colorbar()
```

```python
img_gray=[3]
for i in range(1,3):
    img_gray += img[:,:,i]
img_gray = img_gray/-3
imshow(img_gray, cmap='gray')
```

```python
src_image = img_gray
sx = ndimage.sobel(src_image, axis=0, mode='constant')
sy = ndimage.sobel(src_image, axis=1, mode='constant')

sob = hypot(sx, sy)
imshow(sob)
```

```python
src_image = img
px = ndimage.prewitt(src_image, axis=-1, mode = 'constant')
imshow(px)
colorbar()
```
