```python
# Kurt Kaminski
# MAT201A HW3: Image to Sound via Inverse Fourier Transform
#
# This program produces a sound file from an image by reading
# each row of the image and performing an Inverse Fourier Transform on it. 
# Each IFT is appended sequentially to a one dimensional array until all the 
# rows are processed. 

# In this case, the image used is an aperiodic tiling I previously created in 
# the 3d software Houdini. Specifically, it is a Danzer 7 Fold that has some 
# slight color variation in each tile. It is composed of three shapes that fit
# together to form a pattern with no large scale repetition.
#
# Thanks to David for helping with Fourier transform concepts
#
# http://tilings.math.uni-bielefeld.de/substitution_rules/danzers_7_fold_0
```

```python
# Setup environment

%pylab inline
from __future__ import division
import wave
from scipy.io import wavfile 
rcParams['figure.figsize'] = (10, 5) #wide graphs by default
```

```python
# Load and display image and set up some variables

img = imread('d7f_small.jpg');
height = img.shape[0];
width = img.shape[1];
imshow(img)
print 'Width: {0}, height: {1}'.format(width,height)
```

```python
# Create three arrays with all the pixel information of the image in
# a single dimension, then combine them into a three dimensional array

sigR = [];
sigG = [];
sigB = [];

for each_red_value in img[:,:,0]:
    sigR = concatenate([sigR, each_red_value])
        
for each_green_value in img[:,:,1]:
    sigG = concatenate([sigG, each_green_value])
    
for each_blue_value in img[:,:,2]:
    sigB = concatenate([sigB, each_blue_value])
    
image_array = [];
image_array.append(sigR);
image_array.append(sigG);
image_array.append(sigB);

# Plot the results
suptitle('Pixel Values For Each Channel', fontsize=14)
subplot(311)
ylabel('red')
plot(image_array[0], color='r')
subplot(312)
ylabel('green')
plot(image_array[1], color='g')
subplot(313)
ylabel('blue')
plot(image_array[2], color='b')
```

```python
# A small case demonstrating IFT on a single row of pixels in the red channel

window_length = width;
start = 0
window_list = arange(0, len(image_array[0]), window_length);

window = image_array[0][start : start + window_length]; #read red values from the first 640 pixels
X = fft.irfft(window);
X = X[1:window_length] # skip DC offset (?) and tossing out mirrored ifft results

plot(X)
print 'Max: {0}, Min: {1}'.format(max(X),min(X))
```

```python
# Now I will do another small example showing how I solved hopping.... I'm sure there's a better way

window_size = width;  # one row of pixels in the image
hop_size = int(width*.5);  # hop size is half the sample 
total_size = window_size*3;  # size of empty array, in the full version this will be the total # of pixels
window_list = arange(0, total_size, window_size) # window start and stop points

windowed_array = [0] * total_size;  #fill array with 0's so I can add things in the right spot
non_windowed_array = [0] * total_size;

for i, each_row in enumerate(window_list):
    window = image_array[0][each_row : each_row + window_size]; #grab appropriate window
    # first with no windowing
    X = fft.irfft(window);
    X = X[1:window_size]
    non_windowed_array[each_row : each_row-1+window_size] += X[:window_size];

    # now do windowing
    temp_array = [0] * total_size;
    temp_array[:window_size-1] += X[:window_size];
    temp_array = roll(temp_array, i*hop_size);
    windowed_array += temp_array;
    

subplot(211)
xlabel('No windowing')
plot(non_windowed_array)
subplot(212)
xlabel('Hop size = half sample length')
plot(windowed_array)

# print window_list
```

```python
# Loop through each color channel from earlier and do an IFT on the pixel values, 
# appending the results to a new array 

ifft_array = [];
hop_size = int(width*.5);  # hop size is half the sample 
total_size = height*hop_size;  # size of empty array, due to hop size it will be number of rows * hop size instead of columns 

for i, each_color in enumerate(image_array):
    color_channel = [0] * total_size;
    window_length = width;
    window_list = arange(0, len(image_array[i]), window_length);
    
    for j, each_row in enumerate(window_list):
        window = image_array[i][each_row : each_row + window_length];
        X = fft.irfft(window[0:window_length]);
        X = X[1:window_length]  # skip the first sample, it is always super high. Is this DC offset?
        
        # do the windowing!
        temp_array = [0] * height*hop_size;
        temp_array[:window_size-1] += X[:window_size];
        temp_array = roll(temp_array, j*hop_size);
        color_channel += temp_array;
        
    ifft_array.append(color_channel[:total_size]); # cut off 0's at the end and append color signal to ifft array

# Plot the results
suptitle('Inverse Fourier Transform, Three Color Channels', fontsize=14)
subplot(411)
ylabel('red')
plot(ifft_array[0], color='r')
subplot(412)
ylabel('green')
plot(ifft_array[1], color='g')
subplot(413)
ylabel('green')
plot(ifft_array[2], color='b')
```

```python
# Finally, move each IFT result into the audio spectrum. Special thanks to David for helping
# me with a tidy way of doing this algorithmically.

out_ifft = [];
for i, each_color in enumerate(ifft_array):
    max_val = max(ifft_array[i]);
    min_val = min(ifft_array[i]);
    range = max_val - min_val;

    out_ifft.append( ((ifft_array[i]-min_val)*2)/range - 1 );

# Add them together for a combined signal
out_ifft_combined = (out_ifft[0] + out_ifft[1] + out_ifft[2])/3

# Plot the results
suptitle('Inverse Fourier Transform, Three Color Channels in Audio Spectrum', fontsize=14)
subplot(411)
ylabel('red')
plot(out_ifft[0], color='r')
subplot(412)
ylabel('green')
plot(out_ifft[1], color='g')
subplot(413)
ylabel('green')
plot(out_ifft[2], color='b')
subplot(414)
ylabel('combined')
plot(out_ifft_combined, color='k')

# Write results to wav files
wavfile.write('kaminski_ifft_R.wav', 44100, out_ifft[0]);
wavfile.write('kaminski_ifft_G.wav', 44100, out_ifft[1]);
wavfile.write('kaminski_ifft_B.wav', 44100, out_ifft[2]);
wavfile.write('kaminski_ifft_combined.wav', 44100, out_ifft_combined);
```

```python
# Bonus: multiply the results together for a modulated effect

max_modulated = max(out_ifft[0])*max(out_ifft[1])*max(out_ifft[2]);
out_ifft_modulated = (out_ifft[0] * out_ifft[1] * out_ifft[2])/max_modulated;
plot(out_ifft_modulated)
wavfile.write('kaminski_ifft_combined_modulated.wav', 44100, out_ifft_combined);
```
