```python
# Kurt Kaminski
# MAT201A HW4: Cross-correlation

# In this exercise, I wanted to see what the cross correlation result would be
# across time. I used Eadweard Muybridge's motion study of a horse and rider, cut
# into equal sized jpeg files using Photoshop. The program loads the series of images
# into a dictionary, then cross correlates the first image with the next in the series, then
# the second image with the third, and so on. 
# image source: 
# http://classconnection.s3.amazonaws.com/890/flashcards/1167890/jpg/rp-f-f00634z1333678119361.jpg


# Analysis:

# Muybridge's images provide an interesting comparison for cross correlation. At the beginning of the sequence,
# some large inconsistencies in the background- a black panel moving out of view along frame-left- causes the 
# correlation to produce an inaccurately wide range of higher values in the output matrix. Frames 5-10 are the 
# most consistent in terms of staging, so the cross correlation matrix matches the most in the center where the body
# of the horse is most stable. 

# Perhaps most interestingly, the output matrix matches more closesly along the vertical axis at the head of the sequence
# where the horse is galloping in midair, then transitions into a sharply horizontal match when its feet touch the ground
# and its body elongates.
```

```python
# Setup environment

%pylab inline
from __future__ import division
from scipy.signal import correlate2d 
from scipy.ndimage.filters import maximum_filter
rcParams['figure.figsize'] = (10, 3) #wide graphs by default
```

```python
# load sequence of images into dictionary where img["image #"] = imread("filepath")
img={}
for i in range(1,13):
    img["{0}".format(i)] = imread('muybridge/muybridge_{0}.jpg'.format(i))
    subplot(2,6,"{0}".format(i))
    axis('off')
    subplots_adjust(bottom=.1, top=.9, wspace=.1, hspace=0)
    imshow(img["{0}".format(i)])  
```

```python
# A small example to start. Shows thresholding for the first and second image, then the
# cross correlated result.

i = img["1"]
o = img["2"]
i = sum(i.astype(uint8), axis=-1)/4.0
o = sum(o.astype(uint8), axis=-1)/4.0

i = where(i < 100, 255, 0)
o = where(o < 100, 255, 0)

cc = correlate2d(i, o, mode='same')

subplot(1,3,1)
imshow(i, cmap=cm.gray)
subplot(1,3,2)
imshow(o, cmap=cm.gray)
subplot(1,3,3)
imshow(cc)
```

```python
# Cross correlate all the frames in pairs
cc = {}
for i in range(1,12):
    x = img["{0}".format(i)]
    y = img["{0}".format(i+1)]
    
    x = sum(x.astype(uint8), axis=-1)/4.0
    y = sum(y.astype(uint8), axis=-1)/4.0

    x = where(x < 100, 255, 0)
    y = where(y < 100, 255, 0)
    
    cc["{0}".format(i)] = correlate2d(x, y, mode='same') 
```

```python
# Display the original series of images next to the cross correlated result

for i in range(1,13):
    subplots_adjust(bottom=.1, top=.9, wspace=.1, hspace=.2)
    subplot(4,6,"{0}".format(i))
    imshow(img["{0}".format(i)])
    axis('off')

    if i < 12:
        subplots_adjust(bottom=.1, top=.9, wspace=.1, hspace=.2)
        subplot(4,6,"{0}".format(i+12))
        imshow(cc["{0}".format(i)])
        axis('off')
gcf().set_figwidth(14)
gcf().set_figheight(8)
```
