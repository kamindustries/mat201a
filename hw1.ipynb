```python
%pylab inline
import csv
dateList = []
pizzaList = []
wineList = []
index = []
i = 0
with open('./pizza_wine.csv', 'r') as csvfile:
    #reader = csv.DictReader(csvfile, fieldnames=['date','popularity'])
    reader = csv.reader(csvfile)
    for row in reader:
        dateList.append(row[0][:10])
        pizzaList.append(row[1])
        wineList.append(row[2])
        index.append(i)
        i += 1
        
header1 = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'bold',
        'size'   : 16,
        }
header2 = {'family' : 'sans-serif',
        'color'  : 'black',
        'weight' : 'normal',
        'size'   : 12,
        }
header3 = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'bold',
        'size'   : 12,
        }

suptitle('The "Rise" of Pizza', fontdict=header1)

subplot(111, axisbg='gray')
title("Google search popularity of 'pizza' and 'wine', 2004-2014", fontdict=header2)

ylabel('popularity', fontdict=header3)
xlabel('date', fontdict=header3)
xticks(index[0::52], dateList[0::52], rotation='vertical')
grid()
gcf().set_figwidth(12)
subplots_adjust(top=.85, bottom=-.3)

plot(index, pizzaList, color='gold', linewidth=1.5)
plot(index, wineList, color='darkred', linewidth=1.5)
legend(['pizza','wine'], loc=2, shadow=1)

#source: http://www.google.com/trends/explore#q=pizza%2C%20wine&cmpt=q&tz=

```
